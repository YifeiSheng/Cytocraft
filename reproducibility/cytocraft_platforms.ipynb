```python
import pickle, copy, importlib, math, os
import seaborn as sns
import numpy as np
import pandas as pd
from tqdm import tqdm
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
from cytocraft.craft import *
from cytocraft import model
from cytocraft import util
from cytocraft.model import BasisShapeModel
from cytocraft.plot import *

def plot_RMSD_heatmap(
    D,
    labels,
    cmap="viridis",
    method="average",
    cthresh=4.8,
    font_scale=1,
    cbbox=(0.07, 0.84, 0.04, 0.12),
    lgbox=(0.86, 0.96, 0.01, 0.01),
    figsize=(8,8),
    dpi=300,
):
    clustergrid = sns.clustermap(
        D, xticklabels=labels, yticklabels=labels, method=method
    )
    plt.close()

    # Get clusters with threshold
    linkage = clustergrid.dendrogram_row.linkage
    threshold = cthresh
    clusters = fcluster(linkage, threshold, criterion="distance")

    # Create a color map for clusters
    n_clusters = len(np.unique(clusters))
    color_map = sns.color_palette("rainbow", n_clusters)

    # Assign colors to clusters
    cluster_colors = [color_map[c - 1] for c in clusters]

    # Draw plot
    sns.set(rc={"figure.dpi": dpi, "figure.figsize": figsize}, font_scale=font_scale)

    ax = sns.clustermap(
        D,
        xticklabels=labels,
        yticklabels=labels,
        cmap=cmap,
        method=method,
        row_colors=[cluster_colors],
        col_colors=[cluster_colors],
        cbar_kws={"ticks": [0, 0.5, 1, 1.5]},
    )
    ax.ax_cbar.set_title("NRMSD")
    ax.ax_cbar.set_position(cbbox)

    # Add legend for row color bar
    from matplotlib.patches import Patch

    lut = dict(zip(clusters, cluster_colors))
    handles = [Patch(facecolor=lut[name]) for name in sorted(lut)]
    legend = plt.legend(
        handles,
        sorted(lut),
        title="Clusters",
        bbox_to_anchor=lgbox,
        bbox_transform=plt.gcf().transFigure,
        loc="upper right",
        ncol=2,
    )
    legend.get_frame().set_facecolor("white")
```

```python
def kabsch_numpy(P, Q):
    """
    Computes the optimal rotation and translation to align two sets of points (P -> Q),
    and their RMSD.

    :param P: A Nx3 matrix of points
    :param Q: A Nx3 matrix of points
    :return: A tuple containing the optimal rotation matrix, the optimal
             translation vector, and the RMSD.
    """
    P = np.array(P)
    Q = np.array(Q)
    assert P.shape == Q.shape, "Matrix dimensions must match"

    # Compute centroids
    centroid_P = np.mean(P, axis=0)
    centroid_Q = np.mean(Q, axis=0)

    # Optimal translation
    t = centroid_Q - centroid_P

    # Center the points
    p = P - centroid_P
    q = Q - centroid_Q

    # Compute the covariance matrix
    H = np.dot(p.T, q)

    # SVD
    U, S, Vt = np.linalg.svd(H)

    # Validate right-handed coordinate system
    is_reflection = np.linalg.det(np.dot(Vt.T, U.T)) < 0.0
    if is_reflection:
        Vt[-1, :] *= -1.0

    # Optimal rotation
    R = np.dot(Vt.T, U.T)

    # RMSD
    rmsd = np.sqrt(np.sum(np.square(np.dot(p, R.T) - q)) / P.shape[0])

    return R, t, rmsd, is_reflection
    
def RMSD_distance_matrix(
    adatas, order=None, ngene=100, compare_method="pair", norm_method=None
):
    """
    Calculate the RMSD distance matrix between configurations of different cell types
    Parameters
    ----------
    Confs : dict
        The dictionary of the configurations of different cell types
    GeneLists : dict
        The dictionary of the gene lists of different cell types
    order : list, optional
        The order of the cell types, by default None
    ngene : int, optional
        The number of genes to be compared, by default 100
    compare_method : str, optional
        The method to compare the gene lists, by default "pair"
    norm_method : str, optional
        The method to normalize the configurations, by default None
    Returns
    -------
    np.ndarray
        The RMSD distance matrix
    """
    import numpy as np

    if order == None:
        keys = list(adatas.keys())
    else:
        keys = order

    GeneLists = {}
    for ct, adata in adatas.items():
        if "F" in adata.uns.keys():
            GeneLists[ct] = adata.uns["F"].index
        elif "X" in adata.uns.keys():
            GeneLists[ct] = adata.uns["X"].index   
        else:
            print(f"Please run Cytocraft for {ct} first")
            

    # calculate the distance matrix
    N = len(keys)
    DM = np.zeros((N, N))
    if compare_method == "complete":
        from functools import reduce

        intersected_values = reduce(np.intersect1d, GeneLists.values())
        intersected_values = intersected_values[:ngene]
        if len(intersected_values) < ngene:
            print(
                f"Warning: {len(intersected_values)} common genes between samples are less than {ngene}"
            )
    from tqdm import tqdm

    for n, key_n in enumerate(tqdm(keys)):
        for m, key_m in enumerate(keys[: n + 1]):
            if compare_method == "pair":
                intersected_values = np.intersect1d(GeneLists[key_n], GeneLists[key_m])
                intersected_values = intersected_values[:ngene]
                if len(intersected_values) < ngene:
                    print(
                        f"Warning: {len(intersected_values)} common genes between {key_n} and {key_m} are less than {ngene}"
                    )
            boolean_arrays_n = np.in1d(GeneLists[key_n], intersected_values)
            boolean_arrays_m = np.in1d(GeneLists[key_m], intersected_values)
            if "F" in adata.uns.keys():
                Conf_n = adatas[key_n].uns["F"][boolean_arrays_n]
                Conf_m = adatas[key_m].uns["F"][boolean_arrays_m]
            elif "X" in adata.uns.keys():
                Conf_n = adatas[key_n].uns["X"][boolean_arrays_n]
                Conf_m = adatas[key_m].uns["X"][boolean_arrays_m]
            if norm_method:
                Conf_n = normalizeF(Conf_n, method=norm_method)
                Conf_m = normalizeF(Conf_m, method=norm_method)
            rmsd1 = kabsch_numpy(Conf_n, Conf_m)[2]
            rmsd2 = kabsch_numpy(mirror(Conf_n), Conf_m)[2]
            DM[n, m] = DM[m, n] = min(rmsd1, rmsd2)
    return DM


def plot_RMSD_heatmap(
    D,
    labels,
    cmap="viridis",
    method="average",
    cthresh=4.8,
    font_scale=1,
    legend=True,
    cbbox=(0.07, 0.84, 0.04, 0.12),
    lgbox=(0.86, 0.96, 0.01, 0.01),
    ticks=[0, 0.5, 1, 1.5],
    figsize=(8, 8),
    dpi=300,
    legend_ncol=2,
    mask=None,
):
    import matplotlib.pyplot as plt

    clustergrid = sns.clustermap(
        D, xticklabels=labels, yticklabels=labels, method=method
    )
    plt.close()

    # Get clusters with threshold
    linkage = clustergrid.dendrogram_row.linkage
    threshold = cthresh
    clusters = fcluster(linkage, threshold, criterion="distance")

    # Create a color map for clusters
    n_clusters = len(np.unique(clusters))
    color_map = sns.color_palette("rainbow", n_clusters)

    # Assign colors to clusters
    cluster_colors = [color_map[c - 1] for c in clusters]

    # Draw plot
    sns.set(rc={"figure.dpi": dpi, "figure.figsize": figsize}, font_scale=font_scale)

    ax = sns.clustermap(
        D,
        xticklabels=labels,
        yticklabels=labels,
        cmap=cmap,
        method=method,
        row_colors=[cluster_colors],
        col_colors=[cluster_colors],
        cbar_kws={"ticks": ticks},
        mask=mask,
    )
    ax.ax_cbar.set_title("RMSD")
    ax.ax_cbar.set_position(cbbox)

    if legend:
        # Add legend for row color bar
        from matplotlib.patches import Patch

        lut = dict(zip(clusters, cluster_colors))
        handles = [Patch(facecolor=lut[name]) for name in sorted(lut)]
        legend = plt.legend(
            handles,
            sorted(lut),
            title="Groups",
            bbox_to_anchor=lgbox,
            bbox_transform=plt.gcf().transFigure,
            loc="upper right",
            ncol=legend_ncol,
        )
        legend.get_frame().set_facecolor("white")

    return ax
```

```python
import os
from functools import reduce
import numpy as np
import scanpy as sc
import seaborn as sns
import pandas as pd

Xs={}
GeneLists={}

result_path="/data1/shengyifei/ReST3D/results/Lung/"
samples = [ name for name in sorted(os.listdir(result_path)) if os.path.isdir(os.path.join(result_path, name)) ]
tumorlables = ['Lung5_Rep1_tumor_5', 'Lung5_Rep2_tumor_5', 'Lung5_Rep3_tumor_5','Lung6_tumor_6','Lung9_Rep1_tumor_9','Lung9_Rep2_tumor_9','Lung12_tumor_12','Lung13_tumor_13']
for sample in samples:
    result_dir = os.path.join(result_path,sample)
    files = os.listdir(os.path.join(result_dir))
    info = sample.split(".")
    chip = info[0]
    ctype = info[-1][:-5]
    sample = chip+'_'+ctype
    if 'adata.h5ad' in files and sample in tumorlables: 
        print(sample)
        adata = sc.read_h5ad(os.path.join(result_dir+"/adata.h5ad"))
        Xs[sample] = adata.uns['X']
        GeneLists[sample] = adata.uns['X'].index

intersected_genes = reduce(lambda x, y: x.intersection(y), GeneLists.values())
M = intersected_genes.size
print(M)
```

```python
intersected_genes
```

```python
# Build gene-gene distance matrix
keys = ['Lung5_Rep1_tumor_5', 'Lung5_Rep2_tumor_5', 'Lung5_Rep3_tumor_5','Lung6_tumor_6','Lung9_Rep1_tumor_9','Lung9_Rep2_tumor_9','Lung12_tumor_12','Lung13_tumor_13']

N = len(keys)
GG_dev = np.zeros((N,M,M))
for n in tqdm(range(N)):
    index_k_intersection = [list(GeneLists[keys[n]]).index(gene) for gene in intersected_genes]
    for i, idx_n in enumerate(index_k_intersection):
        for j, idx_m in enumerate(index_k_intersection[:i+1]):
            g1_cord = Xs[keys[n]].iloc[idx_n,:].to_numpy()
            g2_cord = Xs[keys[n]].iloc[idx_m,:].to_numpy()
            d = np.linalg.norm(g1_cord - g2_cord)
            GG_dev[n,i,j] = GG_dev[n,j,i] = d
```

```python
import statistics
V_lung_tumors = np.zeros((M,M))
for i in range(M):
    for j in range(M):
        v = statistics.variance(GG_dev[:,i,j])
        V_lung_tumors[i,j] = V_lung_tumors[j,i] = v
        
V_lung_tumors_sqrt = np.sqrt(V_lung_tumors)
```

```python
# hierarchy
import seaborn as sns
import pandas as pd
import matplotlib.patches as patches
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import fcluster

clustergrid = sns.clustermap(V_lung_tumors_sqrt, xticklabels=intersected_genes, yticklabels=intersected_genes, method='ward')
plt.close()

sns.set(rc={"figure.dpi":300,'figure.figsize':(30, 30)},font_scale=1)

ax = sns.clustermap(V_lung_tumors_sqrt, xticklabels=[], yticklabels=[], cmap="viridis", method='ward', 
    cbar_kws={'ticks': [0,2,4,6,8],'orientation':'horizontal'}, vmax=8)
ax.ax_cbar.set_title("Standard deviation")
ax.ax_cbar.set_position((0.06, .85, .12, .05))


## subconformation 1
# Get clusters with threshold
linkage = clustergrid.dendrogram_row.linkage
threshold = 18
clusters = fcluster(linkage, threshold, criterion='distance')
# add patches
row_order = ax.dendrogram_row.reordered_ind
# start
x = y = 0
# width
w = h = 0
formal_c = 1
count = 1
for c in np.roll(clusters[row_order],-1):
    if c == formal_c:
        formal_c = c
        count += 1
    else:
        w = h = count
        if count>=3:
            ax.ax_heatmap.add_patch(patches.Rectangle((x, y), w, h, fill=False, edgecolor='white', lw=1, clip_on=False))
        x += count
        y = x
        formal_c = c
        count = 1

## subconformation 2
# Get clusters with threshold
linkage = clustergrid.dendrogram_row.linkage
threshold = 9
clusters = fcluster(linkage, threshold, criterion='distance')
# add patches
row_order = ax.dendrogram_row.reordered_ind
# start
x = y = 0
# width
w = h = 0
formal_c = 1
count = 1
for c in np.roll(clusters[row_order],-1):
    if c == formal_c:
        formal_c = c
        count += 1
    else:
        w = h = count
        ax.ax_heatmap.add_patch(patches.Rectangle((x, y), w, h, fill=False, edgecolor='orange', lw=1, clip_on=False))
        x += count
        y = x
        formal_c = c
        count = 1
```

```python
from cytocraft.craft import kabsch_numpy
def str_to_float(fraction_str):
        numerator, denominator = fraction_str.split('/')
        fraction_float = eval(f"{numerator}/{denominator}")
        return fraction_float

def calculate_distances(coords1, coords2):
        # Calculate the Euclidean distance between each pair of coordinates
        distances = np.linalg.norm(coords1 - coords2, axis=1)
        return distances

go_arr=[]
Distances_all = pd.DataFrame(columns=['sample', 'distance'])
for epi_key,tumor_key in zip(['Lung5_Rep1_epithelial', 'Lung5_Rep2_epithelial', 'Lung5_Rep3_epithelial', 'Lung6_epithelial', 
    'Lung9_Rep1_epithelial', 'Lung9_Rep2_epithelial', 'Lung12_epithelial', 'Lung13_epithelial'],['Lung5_Rep1_tumor_5', 
    'Lung5_Rep2_tumor_5', 'Lung5_Rep3_tumor_5', 'Lung6_tumor_6', 'Lung9_Rep1_tumor_9', 'Lung9_Rep2_tumor_9', 'Lung12_tumor_12', 
    'Lung13_tumor_13']):
    sample=epi_key.rstrip('_epithelial')
    conf_epi = Xs[epi_key]
    conf_tumor = Xs[tumor_key]
    intersected_index = GeneLists[epi_key].intersection(GeneLists[tumor_key])
    print(sample+" intersected genes number: "+str(len(intersected_index)))
    conf_epi_subset = conf_epi.loc[intersected_index]
    conf_tumor_subset = conf_tumor.loc[intersected_index]
    R, _, rmsd, reflection = kabsch_numpy(conf_epi_subset, conf_tumor_subset)
    #R, _, rmsd, reflection = kabsch_numpy(conf_epi_subset, conf_tumor_subset)

    # Assuming you have two sets of coordinates stored in variables coords1 and coords2
    coords1 = conf_epi_subset.values
    coords2 = np.dot(conf_tumor_subset.values,R)

    # Calculate the distances and vectors
    distances = calculate_distances(coords1, coords2)
    vectors = coords1 - coords2
    n = len(distances)
    for i in range(n):
        vectors[i,:] /= distances[i]

    # Create a DataFrame with the distances
    distances_df = pd.DataFrame({'sample': sample, 'distance':distances, 'gene':conf_epi_subset.index, 'vector_d1': vectors[:,0], 'vector_d2': vectors[:,1],'vector_d3': vectors[:,2]}, index=conf_epi_subset.index)
    Distances_all = pd.concat([Distances_all,distances_df], ignore_index=True)

    # Sort the distances
    sorted_distances = distances_df.sort_values(by='distance', ascending=False)

    # enrichment analysis
    gene_sets = ['GO_Biological_Process_2023', 'KEGG_2021_Human', 'Reactome_2022']
    result = pd.DataFrame()
    sorted_distances = sorted_distances.reset_index()
    n_genes = len(sorted_distances)
    n_top_genes = int(n_genes * 0.1)
    n_bottom_genes = int(n_genes * 0.9)

    go_arr.append(['Short-distance Translocation Genes '+sample, list(sorted_distances.tail(n_bottom_genes)['index'])])
    go_arr.append(['Long-distance Translocation Genes '+sample, list(sorted_distances.head(n_top_genes)['index'])])
    
for i in range(len(go_arr)):
    name = go_arr[i][0]
    comp = go_arr[i][1]
    enr = gp.enrichr(gene_list=list(comp), # or "./tests/data/gene_list.txt",
                    gene_sets=gene_sets,
                    organism='Human', # don't forget to set organism to the one you desired! e.g. Yeast
                    outdir=None, # don't write to disk
                    )
    res_temp = enr.results
    res_temp['name'] = name
    result_new = res_temp
    #result_new = res_temp[res_temp['Adjusted P-value']<0.05]
    #if result_new.empty:
    #    continue
    result_new['Term'] = result_new['Term'].str.split('(', expand=True)[0]
    result_new['Term'] = result_new['Term'].str.split(' R-HSA', expand=True)[0]
    # result_new = result_new[result_new.loc[:,'Overlap'].apply(str_to_float)>percent]
    result = pd.concat([result,result_new],axis=0, ignore_index=True)

result[['hit', 'pool']] = result['Overlap'].str.split('/', expand=True).astype(int)
result['hit_ratio'] = result['hit'] / result['pool']
result = result[(result['hit'] >= 2)]
```

```python
import numpy as np

def compute_mean_vector(vectors):
    R = np.mean(vectors, axis=0)
    R_length = np.linalg.norm(R)
    return R, R_length

def compute_angular_dispersion(R_length):
    if R_length < 1:
        kappa = R_length / (1 - R_length)
    else:
        kappa = float('inf') # When R_length is 1, kappa is theoretically infinite (perfect alignment)
    return kappa
```

```python
import pandas as pd

# Create an empty DataFrame to store the results
results_df = pd.DataFrame(columns=['Gene', 'Mean Vector', 'Length of Mean Vector (R)', 'Angular Dispersion ()'])

# Loop through each gene
for gene in Distances_filter['gene'].drop_duplicates():
    # Filter the vectors for the current gene
    vectors = Distances_filter[Distances_filter['gene'] == gene][['vector_d1', 'vector_d2', 'vector_d3']].values
    
    # Compute the mean vector and its length
    mean_vector, R_length = compute_mean_vector(vectors)
    
    # Compute the angular dispersion
    kappa = compute_angular_dispersion(R_length)
    
    # Append the results to the DataFrame
    new_row = pd.DataFrame({'Gene': [gene], 'Mean Vector': [mean_vector], 'Length of Mean Vector (R)': [R_length], 'Angular Dispersion ()': [kappa]})
    results_df = pd.concat([results_df, new_row], ignore_index=True)
```

#### circular statistics: 

```python
import numpy as np
import pandas as pd
from scipy.stats import circmean, circstd
from statsmodels.stats.multitest import multipletests
from pycircstat.tests import rayleigh
from pycircstat.descriptive import mean as circ_mean, std as circ_std

def vector_to_angles(vectors):
    # Convert 3D vectors to azimuthal and polar angles
    # azimuthal: angle in xy-plane from x-axis
    # polar: angle from z-axis
    azimuthal = np.arctan2(vectors[:,1], vectors[:,0])
    polar = np.arccos(vectors[:,2] / np.linalg.norm(vectors, axis=1))
    return azimuthal, polar

results = []
genes = Distances_filter['gene'].unique()
for gene in genes:
    vecs = Distances_filter[Distances_filter['gene'] == gene][['vector_d1', 'vector_d2', 'vector_d3']].values
    if len(vecs) < 3:
        continue  # skip genes with too few samples
    azimuthal, polar = vector_to_angles(vecs)
    # Rayleigh test for non-uniformity of azimuthal angles
    stat, pval = rayleigh(azimuthal)
    # Circular mean and std
    mean_dir = circ_mean(azimuthal)
    std_dir = circ_std(azimuthal)
    # Effect size: resultant vector length
    R = np.linalg.norm(np.mean(vecs / np.linalg.norm(vecs, axis=1, keepdims=True), axis=0))
    # Confidence interval for mean direction (approximate, for large n)
    n = len(azimuthal)
    ci_low = mean_dir - 1.96 * std_dir / np.sqrt(n)
    ci_high = mean_dir + 1.96 * std_dir / np.sqrt(n)
    results.append({
        'gene': gene,
        'rayleigh_stat': stat,
        'rayleigh_pval': pval,
        'mean_direction': mean_dir,
        'direction_std': std_dir,
        'effect_size_R': R,
        'mean_direction_CI_low': ci_low,
        'mean_direction_CI_high': ci_high,
        'n_samples': n
    })

results_df = pd.DataFrame(results)
# Multiple testing correction
results_df['pval_adj'] = multipletests(results_df['rayleigh_pval'], method='fdr_bh')[1]
results_df = results_df.sort_values('pval_adj')
results_df.to_csv('circular_stats_rayleigh_results.csv', index=False)
results_df.head(10)
```

```python
import pickle, copy, importlib, math, os
import seaborn as sns
import numpy as np
import pandas as pd
from tqdm import tqdm
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
from cytocraft.craft import *
from cytocraft import model
from cytocraft import util
from cytocraft.model import BasisShapeModel
from cytocraft.plot import *
```

```python
def RMSD_distance_matrix_(Xs, GeneLists, keys, ngene=100, method=None):
    N = len(keys)
    DM = np.zeros((N, N))
    lengene=[]
    for n, key_n in tqdm(enumerate(keys)):
        for m, key_m in enumerate(keys[: n + 1]):
            intersected_values = np.intersect1d(GeneLists[key_n], GeneLists[key_m])
            lengene.append(len(intersected_values))
            intersected_values = intersected_values[:ngene]
            boolean_arrays_n = np.in1d(GeneLists[key_n], intersected_values)
            boolean_arrays_m = np.in1d(GeneLists[key_m], intersected_values)
            X_n = Xs[key_n][boolean_arrays_n]
            X_m = Xs[key_m][boolean_arrays_m]
            if method:
                X_n = normalizeX(X_n, method=method)
                X_m = normalizeX(X_m, method=method)
            rmsd1 = kabsch_numpy(X_n, X_m)[2]
            rmsd2 = kabsch_numpy(mirror(X_n), X_m)[2]
            DM[n, m] = DM[m, n] = min(rmsd1, rmsd2)
    print(np.array(lengene).min())
    return DM

def plot_RMSD_heatmap(
    D,
    labels,
    cmap="viridis",
    method="average",
    cthresh=4.8,
    font_scale=1,
    cbbox=(0.07, 0.84, 0.04, 0.12),
    lgbox=(0.86, 0.96, 0.01, 0.01),
    ticks=[0, 0.5, 1, 1.5],
    figsize=(8, 8),
    dpi=300,
    legend_ncol=2,
    mask=None,
):
    import matplotlib.pyplot as plt

    clustergrid = sns.clustermap(
        D, xticklabels=labels, yticklabels=labels, method=method
    )
    plt.close()

    # Get clusters with threshold
    linkage = clustergrid.dendrogram_row.linkage
    threshold = cthresh
    clusters = fcluster(linkage, threshold, criterion="distance")

    # Create a color map for clusters
    n_clusters = len(np.unique(clusters))
    color_map = sns.color_palette("rainbow", n_clusters)

    # Assign colors to clusters
    cluster_colors = [color_map[c - 1] for c in clusters]

    # Draw plot
    sns.set(rc={"figure.dpi": dpi, "figure.figsize": figsize}, font_scale=font_scale)

    ax = sns.clustermap(
        D,
        xticklabels=labels,
        yticklabels=labels,
        cmap=cmap,
        method=method,
        row_colors=[cluster_colors],
        col_colors=[cluster_colors],
        cbar_kws={"ticks": ticks},
        mask=mask,
    )
    ax.ax_cbar.set_title("RMSD")
    ax.ax_cbar.set_position(cbbox)

    # Add legend for row color bar
    from matplotlib.patches import Patch

    lut = dict(zip(clusters, cluster_colors))
    handles = [Patch(facecolor=lut[name]) for name in sorted(lut)]
    legend = plt.legend(
        handles,
        sorted(lut),
        title="Clusters",
        bbox_to_anchor=lgbox,
        bbox_transform=plt.gcf().transFigure,
        loc="upper right",
        ncol=legend_ncol,
    )
    legend.get_frame().set_facecolor("white")
```

```python
def read_gem_as_csv_v2(path, sep="\t"):
    gem = pd.read_csv(
        path,
        sep=sep,
        comment="#",
        dtype=str,
        converters={
            "x": float,
            "y": float,
            "cell_id": str,
            "cell": str,
            "CellID": str,
            "MIDCount": float,
            "MIDCounts": float,
            "expr": float,
        },
    )
    gem["x"] = pd.to_numeric(gem["x"], errors="coerce").fillna(0).astype(int)
    gem["y"] = pd.to_numeric(gem["y"], errors="coerce").fillna(0).astype(int)
    # cell column
    if "cell" in gem.columns:
        gem.rename(columns={"cell": "CellID"}, inplace=True)
    elif "cell_id" in gem.columns:
        gem.rename(columns={"cell_id": "CellID"}, inplace=True)
    # count column
    if "MIDCounts" in gem.columns:
        gem.rename(columns={"MIDCounts": "MIDCount"}, inplace=True)
    elif "expr" in gem.columns:
        gem.rename(columns={"expr": "MIDCount"}, inplace=True)
    # gene column
    if "gene" in gem.columns:
        gem.rename(columns={"gene": "geneID"}, inplace=True)
    elif "GeneID" in gem.columns:
        gem.rename(columns={"GeneID": "geneID"}, inplace=True)

    return gem
```
